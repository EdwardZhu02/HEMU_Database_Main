## HEMU Database Generator Rev.A1
Parse **CSV-format gene expression matrix** generated from previous steps and combine **Tissue-specific data**
To create comprehensive MySQL database sheets for each of the Andropogoneae species.

这是基于之前生成的CSV文件创建数据库表的脚本。
对于每一个物种，需要创建两个MySQL sheet。其中一个存放Sample对应的基因表达量信息，另一个存放Sample对应的Tissue信息。

```python
import pandas as pd
from sqlalchemy import create_engine
```

### Step1: Preparation
Read gene expression matrix and sample information catalog
and extract **database sheet names**

```python
expression_csv = "./Resources/Zea_mays_expression.csv"
sampleinfo_csv = "./Resources/Zea_mays_sampleinfo.csv"

expression_data = pd.read_csv(expression_csv, low_memory=False, sep=",")
expression_df = pd.DataFrame(expression_data)

sampleinfo_data = pd.read_csv(sampleinfo_csv, low_memory=False, sep=",")
sampleinfo_df = pd.DataFrame(sampleinfo_data)

expression_sheet_name = expression_csv.split("/")[-1].rstrip("_expression.csv").lower() + "_exp"
tissue_sheet_name = expression_csv.split("/")[-1].rstrip("_sampleinfo.csv").lower() + "_spl" # Tissue information of each sample
```

```python
print("Parsed Genes: %s\nParsed Samples: %s" %
      (expression_df.shape[0]-1, int((expression_df.shape[1]-1)/2)))
```
